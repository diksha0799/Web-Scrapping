```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url="http://www.hubertiming.com/results/2017GPTR10K"
html=urlopen(url)
```

```python
print(html)
```

```python
soup=BeautifulSoup(html,'html.parser')
print(soup)
```

```python
# get the title
text=soup.get_text()
print(text)
```

```python
# find hyperlink
all_links=soup.find_all('a')
for link in all_links:
    print(link.get('href'))
```

```python
# print first 10 rows for sanity check
rows=soup.find_all('tr')
print(rows[:10])
```

```python
for row in rows:
    row_td=row.find_all('td')
    print(row_td)
```

```python
str_cells=str(row_td)
cleantext=BeautifulSoup(str_cells,"html.parser").get_text()
print(cleantext)
```

```python
import re
list_rows=[]
for row in rows:
    cells=row.find_all('td')
    str_cells=str(cells)
    clean=re.compile('.</?\w*>')
    clean2=(re.sub(clean,'',str_cells))
    list_rows.append(clean2)
print(list_rows)
    
```

```python
df=pd.DataFrame(list_rows)
df.head(10)
df1=df[0].str.split(',',expand=True)
df1.head(10)
```

```python
df1[0]=df1[0].str.strip('[')
df1[0]=df1[0].str.strip(']')

df1[1]=df1[1].str.strip(']')
df1[1]=df1[1].str.strip('[')
```

```python
col_lables=soup.find_all('th')
col_lables
```

```python
all_header=[]
col_str=str(col_lables)
cleantext2=BeautifulSoup(col_str,"html5lib").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
df2=pd.DataFrame(all_header)
df2
```

```python
df3=df2[0].str.split(',',expand=True)
df3
```

```python
frames=[df3,df1]
df4=pd.concat(frames)
df4.head()
```

```python
df5=df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
df6=df5.dropna(axis=0)
df6
```

```python
df7=df6.drop(df6.index[0])
df7.head()
```

```python
df7.rename(columns={'[Place':'Place'},inplace=True)
df7.rename(columns={'Team]':'Team'},inplace=True)
df7.reset_index(drop=True)
```

```python
time_list=df7[' Chip Time'].tolist()
time_mins= []
s='0'
for i in time_list:
    h=i.split(':')
    math=(int(h[0])*3600+int(h[1])*60+int(s))/60
    time_mins.append(math)
print(time_mins)
```
